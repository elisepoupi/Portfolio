### <a name="1b">Projet d'application</a>
**Équi Caddie !**

*Rappel de l'objectif :*
L'objectif de l'application est de fournir à l'utilisateur, en fonction du contenu de son caddie, son positionnement par rapport aux apports journaliers conseillés et lui proposer des produits complémentaires pour les apports trop faibles, ou des substituts pour les apports trop importants.  
L'application fonctionnera ainsi :  
* l'utilisateur doit au préalable donner les informations suivantes :
 * Années de naissance des membres de la famille
 * Leur sexe
 * s'il est vegan/végétarien
 * s'il souhaite favoriser les produits à faible bilan carbone
 * s'il souhaite favoriser les produits sans additifs
 * s'il souhaite favoriser les produits bio
 * s'il souhaite favoriser les produits sans gluten
 * s'il souhaite favoriser les produits sans trace d'allergène
 * s'il souhaite favoriser les produits sans huile de palme
* l'utilisateur scanne le code-barres de l'article ou des articles
 * l'application lui retourne les données d'apports nutritionnels de ce(s) produit(s) par rapport aux apports journaliers conseillés à partir des quantités par dose données dans la base
* il peut saisir les quantités consommées pour chaque produit (qui remplacent alors les quantités données par la base)
 * l'application lui retourne alors les pourcentages d'apports par rapport aux apports journaliers conseillés
* il peut cliquer sur un apport faible
 * l'application lui retourne quelques produits complémentaires aux apports déjà listés
* il peut cliquer sur un apport trop important
 * l'application lui propose des produits de remplacement aux produits qu'il a saisi, dans des gammes de produits identiques et en prenant en compte ses préférences
 
Maintenant que les données ont été nettoyées, voyons à quoi ressemble notre jeu!

# <a name="2">Analyse</a>
### <a name="2a">Répartition des produits dans les différents groupes</a>

```python
nutri=pd.melt(source,id_vars=['groupe'],value_vars=['fat_100g','proteins_100g','carbohydrates_100g'])
nutri.sort_values(by='groupe',inplace=True)
```

```python
plt.figure(figsize=(11,5))
source.sort_values(by='groupe',inplace=True)
sns.histplot(x=source['groupe'])
plt.xticks(rotation=30,horizontalalignment='right')
plt.xlabel("Groupes")
plt.ylabel('Nombre de produits')
plt.title('Répartition des produits dans les groupes',fontsize=14)
plt.grid(axis='y')
plt.show()
```

```python
x=[]
labels=[]
fig=plt.figure(figsize=(10,4))
for i in nutri['groupe'].unique():
    labels.append(i+' - '+str(round(nutri[nutri['groupe']==i].shape[0]/nutri.shape[0]*100,2))+'%')
    x.append(nutri[nutri['groupe']==i].shape[0])

plt.pie(x,labels=None,autopct='%0.0f%%',startangle=0,pctdistance=1.1)
labels2=['Sans groupe','Avec groupe']
#z=[source.shape[0]-sum(x),sum(x)]

plt.legend(labels,loc=(1.1,0),fontsize=11)
plt.title("Répartition des groupes d'aliments\n après imputation",fontsize=14)

fig.tight_layout()
plt.show()
```

### <a name="2b">Répartition des nutriments, vitamines et minéraux dans les différents groupes</a>

```python
mediane_nutri=source.groupby(by='groupe').median()
mediane_nutri=mediane_nutri.iloc[:,0:26]
mediane_nutri.describe()
```

Pour que ça fonctionne par la suite, nous allons retirer les colonnes pour lesquelles nous n'avons que des valeurs nulles.

```python
mediane_nutri.drop(['pantothenic-acid_100g','vitamin-e_100g','copper_100g','iodine_100g','manganese_100g','selenium_100g','zinc_100g'],axis=1,inplace=True)
```

Il nous faut normaliser les données pour pouvoir les afficher correctement.

```python
mediane_nutri_norm=mediane_nutri.copy()
for i,j in mediane_nutri_norm.iteritems():
    mediane_nutri_norm[i]=mediane_nutri_norm[i]/max(mediane_nutri_norm[i])
#display(mediane_nutri_norm.describe())
#mediane_nutri_norm
```

```python
plt.figure(figsize=(10,5))
plt.subplot(projection='polar')
colors=["#FF0000",'#00D4FF','#FF5000','#0083FF','#FFA100','#0033FF','#FFF200','#1D00FF','#BBFF00','#6E00FF','#6AFF00','#BF00FF','#19FF00','#FF00EE','#00FF37','#FF009D','#00FF87','#FF004C','#00FFD8']
plt.bar(mediane_nutri_norm.columns,mediane_nutri_norm.loc['Foies',:],width=(np.pi/19),color=colors,alpha=0.5)
plt.title('Céréales et féculents',fontsize=14,position=(0.5,1.1))
plt.tick_params(labelbottom=True)
plt.ylim(0,1)
plt.tight_layout()

plt.show()
# pourquoi les étiquettes des variables ne sont-elles pas toutes affichéess?... je ne trouve pas.
# impossible également d'afficher une légende avec les patches de couleurs
```

### <a name="2c">Corrélations entre les différents nutriments</a>

```python
plt.figure(figsize=(12,12))
plt.xticks(fontsize=14)
plt.yticks(fontsize=16)
g=sns.heatmap(source.corr(), cmap="Spectral",annot=False,square=True,vmin=-1,vmax=+1,cbar_kws={'fraction' : 0.045})
g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment='right')
g.collections[0].colorbar.set_label("R", rotation=0,y=1,fontsize=14)
plt.title("Corrélations entre les différentes variables",fontsize=20, pad=30, loc='center')
fig.tight_layout()
plt.show()
```

Nous avons des corrélations entre l'énergie et les lipides, glucides, protides, ce qui est logique puisque l'énergie est calculée à partir de ces 3 macro-nutriments. Je m'attendais à une corrélation plus forte. Sans doute est-ce dû à des données manquantes initialement.  

On a également une corrélation entre les vitamines de la famille B.  

Toujours avec ces vitamines B, nous voyons également de légères corrélations avec le phosphore et le zinc.

Et enfin, plus les produits sont riches en énergie, plus les doses servies sont faibles.
